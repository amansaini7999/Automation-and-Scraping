```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('/home/aman/Documents/Project/WebScrapping/Driver/chromedriver')
driver.get("https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_0_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_0_2_na_na_na&as-pos=0&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=f29c29c7-2226-48dd-a752-9159f9496f35&as-backfill=on")
```

```python
def scrape():
    n = int(driver.find_elements_by_class_name("_2yAnYN")[0].text.split(" ")[3])
    m = int(driver.find_elements_by_class_name("_2yAnYN")[0].text.split(" ")[1])

    for i in range(2,n-m+3):
        name = driver.find_element_by_xpath("//*[@id='container']/div/div[3]/div[2]/div[1]/div[2]/div[" + str(i) + "]/div/div/div/a/div[2]/div[1]/div[1]").text
        price = driver.find_element_by_xpath("/html/body/div[1]/div/div[3]/div[2]/div[1]/div[2]/div[" + str(i) + "]/div/div/div/a/div[2]/div[2]/div[1]/div/div[1]").text
        li.append({"Name" : name, "Price" : price})
```

```python
li = []
noOfPages = int(driver.find_element_by_xpath("//*[@id='container']/div/div[3]/div[2]/div[1]/div[2]/div[26]/div/div/span[1]").text.split(" ")[3])

for j in range(noOfPages):
    scrape()
    if j==0:
        driver.find_element_by_xpath("//*[@id='container']/div/div[3]/div[2]/div[1]/div[2]/div[26]/div/div/nav/a[11]/span").click()
    else:
        driver.find_element_by_xpath("//*[@id='container']/div/div[3]/div[2]/div[1]/div[2]/div[26]/div/div/nav/a[12]/span").click()
    time.sleep(3)
```
